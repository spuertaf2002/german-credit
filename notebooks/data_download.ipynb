### Questions
#### What is the objective of the problem?
To tell if a given credit request is good (low risk) or bad (high risk) for a bank. 
#### How will your solution be used?
- Help to determine most important variables for a low or high risk credit.
- Assist banks for discovering if a given credit request is either high or low on risk to the bank.
- Determine if a credit requested is given or not. 
#### What are the current solutions (if any)?
I did a quick research and found out that some banks currently use **S**upport **V**ector **M**achines for this labour.
#### How should this problem be framed (supervised/unsupervised, online/offline, etc.)
- Problem type: Supervised, Data is labeled and has a target column.
- Learning type: Classification, We'd like to know if a quiven credit request is risky or not.
- Mode of use: Off-line or On-line dependes on the bussiness case scenario.
#### How should the performance of the solution be measured, a first intuition?
The model performance evaluation metric would depend aswell on the bussiness case scenario and priorities.
- If it's more important to evict high risk credits, **recall** should be the metric to prioritize.
- If it's more important to evict denials of credits for good client, **precision** should be the metric to prioritize.
- If a balance between the two aspects mentioned above it's needed, **F1-score** should be the metric to prioritize.
#### Is the performance measure aligned with the objective of the problem?
Yes, the selected performance measures (Recall, Precision, or F1-score) align with the problem's objective. Since the goal is to classify credit requests into high-risk or low-risk categories, it's crucial to balance false positives (wrongly rejecting good clients) and false negatives (approving risky clients).

- If the primary goal is avoiding high-risk approvals, recall is appropriate.
- If the priority is minimizing the rejection of good clients, precision should be the focus.
- If both are equally important, F1-score is the best metric.

Thus, the chosen performance measure depends on the business priorities, but the overall approach is well-aligned with the problem's objective.
#### What would be the minimum performance or yield needed to achieve the objective of the problem?
For the model to be usefull:
- Some financial institutions might require at least 80-90% recall for high-risk detection to reduce defaults.
- AUC-ROC above 0.8 is often considered a good threshold in risk classification.
#### What are the similar problems? Can experiences or tools already created be reused?
Similar problems include:
- Fraud detection: Similar to credit risk classification, fraud detection models aim to classify transactions as fraudulent or legitimate. Techniques like decision trees, logistic regression, and neural networks are often used.
- Loan default prediction: Banks and financial institutions use machine learning to predict whether a borrower will default on a loan. This problem shares similar features, such as customer credit history, income, and transaction patterns.
- Customer churn prediction: Similar in that it requires identifying at-risk customers based on past behavior patterns.
#### Is there experience of the problem available?
- Traditional models: Logistic Regression, Decision Trees, Support Vector Machines (SVM)
- Advanced models: Gradient Boosting Machines (XGBoost, LightGBM), Neural Networks
- Regulatory frameworks: Basel Accords (Basel II, Basel III) provide guidelines for risk modeling in banking
- Publicly available datasets: Kaggle, UCI Machine Learning Repository, and financial institutions often provide anonymized datasets

#### List the assumptions that exist at this time.
- Data is available and labeled: The dataset includes enough labeled examples of high- and low-risk credit requests.
- Features are meaningful: Variables like income, credit score, and repayment history contribute to risk classification.
